# Cereal Nutrition Data 
### 80 Cereal Products

---

**Fields in the dataset:**

* Name: Name of cereal
* mfr: Manufacturer of cereal
  * A = American Home Food Products;
  * G = General Mills
  * K = Kelloggs
  * N = Nabisco
  * P = Post
  * Q = Quaker Oats
  * R = Ralston Purina
* type:
  * cold
  * hot
* calories: calories per serving
* protein: grams of protein
* fat: grams of fat
* sodium: milligrams of sodium
* fiber: grams of dietary fiber
* carbo: grams of complex carbohydrates
* sugars: grams of sugars
* potass: milligrams of potassium
* vitamins: vitamins and minerals - 0, 25, or 100, indicating the typical percentage of FDA recommended
* shelf: display shelf (1, 2, or 3, counting from the floor)
* weight: weight in ounces of one serving
* cups: number of cups in one serving
* rating: a rating of the cereals (Possibly from Consumer Reports?)

### Plots

 * **Fiber and Calories**

```python
df.plot.scatter(x="calories", y="fiber")
plt.xlabel('calories')
plt.ylabel('fiber')
plt.title('Comparing Calories and Fiber content')
plt.show()
```

 * **Fiber and Vitamin**

```python
df.plot.scatter(x="vitamins", y="fiber")
plt.xlabel('vitamins')
plt.ylabel('fiber')
plt.title('Comparing Fiber and Vitamin content')
plt.show()
```

 * **Comparing Nutrients Top Rated**

```python
fiber = [14.0, 4.0, 3.0]
sugars = [0, 0, 0]
carbo = [8.0, 19.0, 20.0]
  
n=3
r = np.arange(n)
width = 0.25

from matplotlib.pyplot import figure
figure(figsize=(8, 6), dpi=80)
  
plt.bar(r, fiber, color = 'b',
        width = width, edgecolor = 'black',
        label='fiber')
plt.bar(r + width, sugars, color = 'g',
        width = width, edgecolor = 'black',
        label='sugars')
plt.bar(r + width, carbo, color = 'r',
        width = width, edgecolor = 'black',
        label='carbo')
  
plt.xlabel("")
plt.ylabel("")
plt.title("Comparing Top 3 Rated")
  
# plt.grid(linestyle='--')
plt.xticks(r + width/2,['All-Bran with Extra Fiber',"Shredded Wheat 'n'Bran",'Shredded Wheat spoon size'])
plt.legend()
  
plt.show()
```

**Note:** Sugar content is zero
